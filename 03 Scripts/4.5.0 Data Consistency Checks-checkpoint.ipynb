```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Import wrangled orders dataset csv files as df_ords_1

df_ords_1 = pd.read_csv(r'C:\Users\IDONG\Prepared data\orders_wrangled.csv', index_col = False)
```

```python
# First things first, always good to get a rough idea of the data set

df_ords_1.head()
```

## Checking for mixed-type data

```python
for col in df_ords_1.columns.tolist():
  weird = (df_ords_1[[col]].applymap(type) != df_ords_1[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_ords_1[weird]) > 0:
    print (col)
```

```python
# Great! No detected mixed type data. Next step is checking for missing values
```

## Checking for missing values

```python
df_ords_1.isnull().sum()
```

```python
# Quite interesting, a whooping 206,209 missing values in the 'days_since-prior_order' column has been detected.
# But observing the output from 'df_ords_1.head()', the first value in the 'days_since-prior_order' column is represented as NaN
# This is because the first row is representative of the first order by user_id '1', thus there cannot be any initial order
# preceeding the first. This is why the value is represented as NaN
# Observing the tail end of the dataset below, notice the last user_id is 206,209. Thus each user would have 'NaN' in its first
# row, which is its first order. 206,209 users transaltes to 206209 'missing values'

df_ords_1.tail()
```

```python
# RECOMMENDATION

# Removing, replacement or calculation of a mean value for 'NaN' would not be necessary at this stage (or even subsequently).
# Apart from fully understanding the reason/meaning for NaN, removing first order rows of all 206209 users from the dataset
# would be loosing vital information. This is because these rows are representative of the onboarding of new users/clients.
# Thus they are bound to contain certain characteristics for predicting possible trends of other new or potential users.

# Alternatively, we can open a subset for the NaN values for special observation. Recall they are like freshly onboarded clients
# Extracting these variables will make direct observation easier 
# Order number 1 was used to create this subset because it is the first order for every user

df_ords_1[df_ords_1['order_number']==1]
```
