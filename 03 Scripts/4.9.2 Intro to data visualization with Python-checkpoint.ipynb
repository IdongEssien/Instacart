##### 1.4 Creating a Line Chart (from sample subset)

```python
# First the subset of the dataframe is generated as df_subset. (For the first 2000000 rows)

df_subset = df_visual[:2000000]
```

```python
# Commands below are the inital steps taken when trying to create a line chart by sampling from a dataframe
# The first command simple ensures that the random distribution is reproducible
# 'dev' ensures selected values are allocated to the appropriate sample. In this case, its either allocated to the big or small
# sample depending on whether it exceeds 0.7 or not
# 'np.random.rand()' is the function that does the random selection process from the dataset specified in the parentheis

np.random.seed(4)
dev = np.random.rand(len(df_subset)) <= 0.7
```

```python
# 'big' represents the larger sample (70% of the sample from df_subset)

big = df_subset[dev]
```

```python
# 'small' represents the smaller sample (30% of the sample from df_subset)

small = df_subset[~dev]
```

```python
# Reconfirmation to ascertain the total sum each sample length is equal to that of the subset (2000000 rows)

len(big) + len(small)
```

```python
# Creation of the line chart for the smaller sample. 'df_small' represents the smaller sample dataset

df_small = small[['orders_day_of_week','prices']]
```

```python
# 'line' is used to save the line chart created from the smaller sample

line = sns.lineplot(data = df_small, x = 'orders_day_of_week', y = 'prices')
```

```python
# A check is also done on the larger sample to confirm similarity of the line chart appearance

df_big = big[['orders_day_of_week','prices']]
```

```python
line_2 = sns.lineplot(data = df_big, x = 'orders_day_of_week', y = 'prices')
```

##### 2.2 Marketing team curious about distribution of orders among customers (A bar chart from the 'loyalty_flag' column will be presented)

```python
# The loyalty_flag column was removed to create memory space for the merge in part 1 of this task. So it will be recreated

# Starting by introducing a 'max_order' column

df_visual['max_order'] = df_visual.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Then allocating the customer category according to the maximum number of orders made by custoemrs.
# For this, the 'loyalty_flag' column is recreated

df_visual.loc[df_visual['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_visual.loc[(df_visual['max_order'] <=40) & (df_visual['max_order'] >10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_visual.loc[df_visual['max_order'] <=10, 'loyalty_flag'] = 'New customer'
```

```python
# A quick reconfirmation of the allocation

df_visual['loyalty_flag'].value_counts(dropna = False)
```

```python
# Now to create the bar chart proper. 'bar_2' will represent the created chart

bar_2 = df_visual['loyalty_flag'].value_counts().plot.bar()
```

##### 2.5 Determination of the connection between age and income (using a scatterplot)

```python
# 'scatplot_1' represents the resultant scatterplot chart

scatplot_1 = sns.scatterplot(x = 'age', y = 'income', data = df_visual)
```

##### Observation - We have 3 categories of customer density from the scatterplot. (Category 1) - Customers occupying the lower end of the chart earning below 200,000. This category is the most dense and it's regardless of the age. (Category 2) - This group is less dense than the first category, earning between 200,000 and 300,000 with ages of 40 and above. (Category 3) - The third category has a scanty appearance in distribution and occupy two portions in the chart. The customers below 40 years who earn between 200,000 and 400,000 occupying one region, while the second set earning above 300,000 and are above 40years occupy the second scanty distributed region.
##### **In summary, there is no clear cut and distinct relation between spending power and age, just some level of segmentation.
