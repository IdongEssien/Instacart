##### 1.1 Recreating a spending flag (to determine spending habits among regions)

```python
# The spending flag had to be removed in earlier task for the sake of memory space and efficiency. It will be recreated
# First we establish the average price spent per user on orders. We do this by creating a new column 'av_price' and at the same
# time, using the syntax below to compute the mean values 

df_insta['av_price'] = df_insta.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Now to categorize according to the requirements of our Instacart clients into 'low & high spenders'.
# Average price below $10 is a low spender; Equal or above $10 is a high spender

df_insta.loc[df_insta['av_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_insta.loc[df_insta['av_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Doing a quick check at the allocation to the 'spending_flag' column

df_insta['spending_flag'].value_counts(dropna = False)
```

```python
# Now to cross reference this spending habit with customers of the 4 regions. 'spend_hab will represent the cross-referencing'

spend_hab = pd.crosstab(df_insta['region'], df_insta['spending_flag'], dropna = False)
```

## 2.0 Detecting customers with low revenue inflow

```python
# The client has requested to identify and exclude customers with a maximum of below 5 orders
# We first determine the maximum number of orders for each user/customer with the syntax below

df_insta['max_order'] = df_insta.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# 'Low activity' will represent the customers to be excluded.

df_insta.loc[df_insta['max_order'] < 5, 'exclusion_flag'] = 'Low activity'
```

```python
# While 'Active' are the customers the client is interested in

df_insta.loc[df_insta['max_order'] >= 5, 'exclusion_flag'] = 'Active'
```

##### 3.1 Aggregate position of max, mean and min for app usage frequency and expenditure

```python
# We can readily define customer app usage frequency according to the days inbetween orders per customer.
# We achieve this using the 'days_since_prior_order'. (Recall it was removed initially for sake of efficiency and memory space).
# To do this, we calculate the median in the 'days_since_prior_order' in relation to the 'user_id'.
# We still make use of the groupby()function. 'ord_freq' is the new column created which serves as order frequency
# Non-frequent customer: when value for 'days_since_prior_order' is higher than 20
# Regular customer: when value is more than 10 but lower or equal to 20
# Frequent customer when value is equal to or less than 10

df_new['ord_freq'] = df_new.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# 'úsage_freq' is the calculated median value per customer 

df_new.loc[df_new['ord_freq'] > 20, 'usage_freq'] = 'Non frequent customer'
```

```python
df_new.loc[(df_new['ord_freq'] <=20) & (df_new['ord_freq'] >10), 'usage_freq'] = 'Regular customer'
```

```python
df_new.loc[df_new['ord_freq'] <= 10, 'usage_freq'] = 'Frequent customer'
```

```python
# Checking for distribution pattern

df_new['usage_freq'].value_counts(dropna = False)
```

```python
# Groupng frequency app usage vs the mean, min and max expenditure

df_new.groupby('usage_freq').agg({'prices': ['mean', 'min', 'max']})

# All 3 categories follow the same average expenditure pattern from the output below
```
